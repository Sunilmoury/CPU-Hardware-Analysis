# cpu vs speed 

```python
ccs=x.sort_values(by="Speed(GHz)", ascending=False)
```

```python
plt.figure(figsize=(16,10))
sns.barplot(data=ccs,x="Speed(GHz)",y="PRODUCT_NAME",color="red")
plt.title("SPEED(GHz) VS CPU",c="r",size="15")
plt.xlabel("Speed(GHz)",c="BLUE",size="15")
plt.ylabel("CPU",c="BLUE",size="15")
plt.grid(ls=":",)
```

# ------------------x----------------

# core vs speed 

```python
plt.figure(figsize=(16,10))
sns.regplot(data=ccs,x="Cores",y="Speed(GHz)",color="crimson")
plt.title("Cores VS SPEED(GHz)",c="r",size="15")
plt.xlabel("Cores",c="BLUE",size="15")
plt.ylabel("SPEED(GHz)",c="BLUE",size="15")
plt.grid(ls=":",)
```

# SALES OF CPU IN DIFFERENT COUNTRIES

```python
pvs=x.groupby("PRODUCT_NAME")
```

```python
A=pvs.get_group("Intel Xeon E5-1650 V4")
```

```python
B=pvs.get_group("Intel Core i7-4790K")
```

```python
C=pvs.get_group("Intel Core i7-7820X")
```

```python
D=pvs.get_group("Intel Core i9-7900X")
```

```python
E=pvs.get_group("Intel Core i7-5930K")
```

```python
cc=pd.concat([A,B,C,D,E])
cc
```

```python
plt.figure(figsize=(14,8))
sns.scatterplot(data=cc,x="COUNTRY_NAME",y="LIST_PRICE",hue="PRODUCT_NAME")
plt.grid(ls=":")
plt.title("PRICES OF CPU VS COUNTRIES",c="r",size="15")
plt.xlabel("COUNTRIES",c="BLUE",size="15")
plt.ylabel("LIST_PRICE",c="BLUE",size="15")
plt.show()
```
